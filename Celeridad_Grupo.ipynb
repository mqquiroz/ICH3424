```python
#Importamos Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
letra = 12.5
alto  = 5
ancho = 10
fig = plt.figure(figsize=(ancho, alto))

k0 = 1/10
dk = 0.01
A = 10.0

x_Cgt = np.arange(-2*2*np.pi/dk,2*2*np.pi/dk,1)

env = 2*A*np.sin(dk*x_Cgt)/x_Cgt

eta = env*np.exp(1j*(k0*x_Cgt))

plt.plot(x_Cgt,np.real(eta),'b', label='Celeridad de Fase $C$ ')
plt.plot(x_Cgt,env,'k',label='Celeridad de Grupo $C_{g}$')
plt.plot(x_Cgt,-env,'k')
plt.xlim(-2*2*np.pi/dk,2*2*np.pi/dk)
plt.ylabel('Amplitud',fontsize=letra)
plt.xlabel('$(x-Ct)$',fontsize=letra)
plt.legend(fontsize=letra-2.5)
plt.grid(which='both',color='lightgray', linestyle='--', linewidth=0.25)
plt.minorticks_on()
plt.savefig('WavePacket.pdf',format='pdf',dpi=300)
plt.show()
```
